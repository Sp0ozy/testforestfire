```python
df[['index', 'day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['index', 'day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
# change the required columns to float data type
objects=[features for features in df.columns if df[features].dtype=='object']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```
